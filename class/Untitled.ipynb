```python
import numpy as np;
import matplotlib.pyplot as plt;
from PIL import Image
from numpy import ndarray as array;

def svd(matrix, k):
    #奇异值分解
    u, z, v = np.linalg.svd(matrix);
    #取前k个特征
    u = u[:, 0:k];
    z = np.diag(z[0:k]);
    v = v[0:k, :];
    #重新组合图片
    a = u.dot(z).dot(v);
    #把不正常的颜色变成正常 !important
    a[a < 0] = 0;
    a[a > 255] = 255;
    matrix[:,:] = a;

image:array = np.array(Image.open("OIP.jpg"));
svd(image[:,:,0], k1);
svd(image[:,:,1], k1);
svd(image[:,:,2], k1);
plt.title("k={}".format(k1))

plt.imshow(image);
plt.show();
```
